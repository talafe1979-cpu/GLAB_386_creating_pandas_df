# Creating a pandas Data Frame

```python
import numpy as np
import pandas as pd
```

```python
d = {'x': [1, 2, 3], 'y': [2, 4, 8], 'z': 100}

df = pd.DataFrame(d)
```

```python
df2 = pd.DataFrame(d, index=[100, 200, 300], columns=['z', 'y', 'x'])

df2
```

```python
student_dict = {
    'Name': ['Joe', 'Nat', 'Harry'],
    'Age': [20, 21, 19],
    'Marks': [85.10, 77.80, 91.54]
}

student_df = pd.DataFrame(student_dict)

student_df.head()
student_df.dtypes
```

```python
student_df = student_df.set_index('Name')

print(student_df.head())
```

```python
data = {
    'Product': ['Laptop', 'Tablet', 'Phone'],
    'Price': [1200, 300, 800],
    'Quantity': [50, 150, 100]
}
df3 = pd.DataFrame(data)

df3.head()
```

```python
df3.index = pd.RangeIndex(start=1, stop=4, step=1)

df3.head()
```

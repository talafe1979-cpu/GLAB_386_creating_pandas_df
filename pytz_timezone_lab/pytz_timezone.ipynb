# Time Zone Conversion with Pytz


```python
import pytz
from datetime import datetime
```

```python
def convert_time_zone(dt, from_tz, to_tz):

    from_tz = pytz.timezone(from_tz)
    to_tz = pytz.timezone(to_tz)

    dt = dt.replace(tzinfo=from_tz)

    dt = dt.astimezone(to_tz)

    return dt
```

```python
today_datetime = datetime.now()

from_timezone = 'US/Eastern'
to_timezone = 'Australia/Melbourne'
```

```python
converted_time = convert_time_zone(today_datetime, from_timezone, to_timezone)
print(converted_time)
```
